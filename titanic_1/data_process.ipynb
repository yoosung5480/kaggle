```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

test.head()
```

```python
## 전처리 대상데이터
# sex -> 여성 : 0, 남자 :1
# name -> 명칭만 뽑은후, 카테고리화
# embarked -> 선박장 별로 카테고리화

# cabin -> 결측값이 너무 많음으로 제거
# tiket -> 연관성이 없음으로 제거
data = train
```

```python
# sex -> 여성 : 0, 남자 :1
encoded_sex = pd.Series([1 if i == 'male' else 0 for i in data['Sex']])
data['Sex'] = encoded_sex
data.head()
```

```python
# name -> 명칭만 뽑은후, 카테고리화
import re
title_dict = {
    'Mr' : 1, 
    'Miss' : 2, 
    'Mrs' : 3,
    'Master' : 4, 
    'Dr' : 5, 
    'Rev' : 6
}

def extract_title(name):
    title = re.search(r'\b(\w+)\.',name)
    return title.group(1)


def encode_title(title):
    if title in title_dict.keys():
        return title_dict[title]
    else:
        return 0
        


title = data['Name'].apply(extract_title)
encoded_title = title.apply(encode_title)
data['Name'] = encoded_title
data.head()
```

```python
encoded_embarked = data['Embarked'].astype('category').cat.codes
data['Embarked'] = encoded_embarked
data.head()
```

```python
x_data = []
for i, rows in data.iterrows():
    x_data.append([rows['Survived'], rows['Pclass'], rows['Name'], rows['Sex'], rows['Age'], rows['SibSp'] ,rows['Parch'], rows['Fare'], rows['Embarked']] )

x_data = pd.DataFrame(x_data)
x_data.head()
```

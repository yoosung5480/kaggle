```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

data = pd.read_csv('train.csv')
data.head()
```

```python
## sex - encoding
encoded_sex = pd.Series([1 if i == 'male' else 0 for i in data['Sex']]) # 간단한 파이썬 제네레이터를 통해 인코딩 방법
data['Sex'] = encoded_sex
sex_count = encoded_sex.value_counts(normalize= True) # 데이터프레임 내에서 각 수치의 빈도, 도는 비율을 계산해주는 함수. value_counts
plt.pie(sex_count, labels = ['Male', 'Female'])
plt.show()
```

```python
## Cabin - encoding
processed_cabin = data['Cabin'].fillna('N').apply(lambda x : x[0])
encoded_cabin = processed_cabin.astype('category')
data['Cabin'] = encoded_cabin
data['Survived'] = encoded_sex

survival_data = data[data['Survived'] > 0]
data
```

```python
# 전체 인원에서의 객실별의 인원 비율
Cabin_ratio = data['Cabin'].value_counts(normalize=True).sort_index()
plt.bar(Cabin_ratio.index, Cabin_ratio.values)
plt.show()
```

```python
# 객실별 생존자의 수
survival_counts = data[data['Survived'] == 1]['Cabin'].value_counts(normalize=False).sort_index()
plt.bar(survival_counts.index, survival_counts.values)
plt.show()
```

```python
# 객실별 사망자의 수
survival_counts = data[data['Survived'] == 0]['Cabin'].value_counts(normalize=False).sort_index()
plt.bar(survival_counts.index, survival_counts.values)
plt.show()
## 결론
# N인 인원이 제일 많지만, 사망과의 어떤연관성도 많이보이지 않고, 일단 중요한건 결측값이란것이다. 이는 머신러닝할때 제외해야할값으로 생각된다.
```

```python
embarked = data[data['Survived'] == 1]['Embarked'].value_counts(normalize=False)
# plt.pie(embarked, labels= ['S', 'C', 'Q'])
plt.bar(embarked.index, embarked.values)
plt.show()
```
